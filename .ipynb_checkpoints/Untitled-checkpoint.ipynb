```python
import numpy as np
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
with open('coursera_sessions_train.txt', 'r') as f:
    train = f.read().splitlines()
with open('coursera_sessions_test.txt', 'r') as f:
    test = f.read().splitlines()
```

```python
train
```

```python
def list_maker(df):
    lst_=[]
    for i in df:
        look_, purchase_ = i.split(';')
        look_= map(int, look_.split(','))
        if len(purchase_) > 0:
            purchase_ = map(int, purchase_.split(','))
        else:
            purchase_ = []
        lst_.append([look_, purchase_])
    return lst_
```

```python
train_mappin=list_maker(train)
test_mappin=list_maker(test)
```

```python
train_look = [row[0] for row in train_mappin]
train_look_numpy = np.array( [i for j in train_look for i in j] )
train_look_counted = np.transpose(np.unique(train_look_numpy, return_counts=True))
train_look_counted = train_look_counted[train_look_counted[:,1].argsort()][::-1]
```

```python
train_look_counted #looks
```

```python
train_purch = [row[1] for row in train_mappin]
train_purch_numpy = np.array( [i for j in train_purch for i in j] )
train_purch_counted = np.transpose(np.unique(train_purch_numpy, return_counts=True))
train_purch_counted = train_purch_counted[train_purch_counted[:,1].argsort()][::-1]
```

```python
train_purch_counted #purchases
```

```python
def metrics(s, r, k):
    purchase = 0
    for ind in r:
        if ind in s:
            purchase += 1 
    precision = purchase / k
    recall = purchase / len(s)
    return(precision, recall)
```

```python
prec_at_1_tr_l, rec_at_1_tr_l = [], []
prec_at_5_tr_l, rec_at_5_tr_l = [], []
k1, k5 = 1, 5
for i, p in enumerate(train_purch):
    # skip sessions without purchases
    if p == []: continue
    
    # looks ids
    l = train_look_counted[i]

    # sorted looks ids indices in sess_train_l_cnt array
    # sort in accordance with looks counts
    l_ind_sess = []
    for j in range(len(l_ind_sess)):
        l_i.append(np.where(train_look_counted[:,0] == l[j])[0][0])
    l_i_sorted = np.unique(l)
    
    # k1 recommendations
    num = min(k1, len(l))
    if num == 0: continue
    recs_k1 = train_look_counted[l_i_sorted[:num],0]
    
    # k1 metrics
    prec_1, rec_1 = metrics(p, recs_k1, k1)
    prec_at_1_tr_l.append(prec_1)
    rec_at_1_tr_l.append(rec_1)
    
    # k5 recommendations
    num_of_recs_k5 = min(k5, len(l))
    if num_of_recs_k5 == 0: continue
    recs_k5 = train_look_counted[l_i_sorted[:num_of_recs_k5],0]
    
    # k5 metrics
    prec_5, rec_5 = metrics(p, recs_k5, k5)
    prec_at_5_tr_l.append(prec_5)
    rec_at_5_tr_l.append(rec_5)
```
